Graphs & Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Histogram

```python
plt.hist(df["tenure"])
plt.title("histogram for tenure")
plt.xlabel("tenure")
plt.ylabel("frequency")
plt.show()
```

CountPlot

```python
sns.countplot(x=df["PaymentMethod"],data=df)
plt.title("count plot for paymenymethod")
plt.xlabel("payment method")
plt.show()
```

```python
sns.countplot(x=df["gender"],data=df)
plt.title("count plot for gender")
plt.xlabel("gender")
plt.show()
```

```python
sns.countplot(x=df["MultipleLines"],hue="InternetService",data=df)
plt.title("multipleline wise internet service")
plt.xlabel("multiple lines")
plt.ylabel("internet service")
plt.show()
```

```python
sns.countplot(x=df["gender"],hue="Partner",data=df)
plt.title("gender wise partner")
plt.xlabel("gender")
plt.ylabel("partner")
plt.show()
```

```python
sns.countplot(x=df["PaymentMethod"],hue="Churn",data=df)
plt.title("PaymentMethod wise churn")
plt.xlabel("PaymentMethod")
plt.ylabel("churn")
plt.xticks(rotation=80)
plt.show()
```

Scatter Plot

```python
plt.scatter(df["MonthlyCharges"],df["tenure"])
plt.title("scatter plot for monthly charges and tenure")
plt.xlabel("monthly charges")
plt.ylabel("tenure")
plt.show()
```

```python
plt.scatter(df["StreamingTV"],df["tenure"])
plt.title("scatter plot for StreamingTV and tenure")
plt.xlabel("StreamingTV")
plt.ylabel("tenure")
plt.show()
```

```python
sns.scatterplot(x="SeniorCitizen",y="tenure",hue="gender",data=df)
plt.title("scatter plot for senior citizen and monthly charges ")
plt.show()
```

Box Plot

```python
sns.boxplot(x="MonthlyCharges",y="Churn",data=df)
plt.title("box plot for monthly charges to churn")
plt.xlabel("monthly charges")
plt.ylabel("churn")
plt.show()
```

```python
sns.boxplot(x="tenure",y="Churn",data=df)
plt.title("box plot for monthly charges to churn")
plt.xlabel("tenure")
plt.ylabel("churn")
plt.show()
```

```python
sns.boxplot(x="Churn",y="SeniorCitizen",hue="tenure",data=df)
plt.title("box plot for churn and senior citizen")
plt.show()
```

Reggreation Plot

```python
sns.regplot(x="tenure",y="MonthlyCharges",data=df)
plt.title("regplot for tenure to monthlucharges")
plt.show()
```

Pair Plot

```python
sns.pairplot(df[["MonthlyCharges","tenure","SeniorCitizen"]])
plt.show()
```

Cross Tab

```python
ct=pd.crosstab(df["gender"],df["Partner"])
ct
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(ct,annot=True,cmap="Blues")
plt.title("gender usage by partner")
plt.xlabel("gender")
plt.ylabel("partner")
plt.show()
```

```python
ct=pd.crosstab(df["PaymentMethod"],df["Churn"])
ct
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(ct,annot=True,cmap="Blues")
plt.title("payment methodr usage by churn")
plt.xlabel("payment method")
plt.ylabel("churn")
plt.show()
```
