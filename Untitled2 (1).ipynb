```python
# 1......wrie a python program that take a password as input and checks if it meets the following reqirments...
p=input("enter password")
count=0
uppercase=False
lowercase=False
digit=False
for i in p:
    count+=1
    if i.isupper():
        uppercase=True
    elif i.islower():
        lowercase=True
    elif i.isdigit():
        digit=True
if count>=8 and uppercase and lowercase and digit:
    print("password is valid")
else:
    print("password is invalid")
```
