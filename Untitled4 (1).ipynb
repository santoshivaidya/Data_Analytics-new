```python
# 24 rearrainging string so lowercase letter appear first. then uppercase.........
s=input("enter string")
lowercase=""
uppercase=""
for i in s:
    if i.islower():
        lowercase+=i
    elif i.isupper():
        uppercase+=i
result=lowercase + uppercase
print(result)
```
