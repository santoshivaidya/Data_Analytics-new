Encoding

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
gender=df["gender"]
encoder=LabelEncoder()
gender_encoded=encoder.fit_transform(gender)
print(gender_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
cdow=df["claim_day_of_week"]
encoder=LabelEncoder()
cdow_encoded=encoder.fit_transform(cdow)
print(cdow_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
it=df["incident_type"]
encoder=LabelEncoder()
it_encoded=encoder.fit_transform(it)
print(it_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```
