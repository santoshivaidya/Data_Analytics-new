```python
# label encoding 
from sklearn.preprocessing import LabelEncoder
flowers = ["rose","lily","sunflower","daisy","tuli","orchid","carnation"]
encoder = LabelEncoder() # initializing label encoding
flowers_encoded = encoder.fit_transform(flowers) # apply it on the target column using fit_transform method
print(flowers_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
# one hot encoding using pandas
import pandas as pd
data={
    "student":["aa","bb","cc","dd"],
    "grades":["A","A+","B","A"],
    "course":["bca","bsc","bba","bca"],
    "city":["pune","mumbai","nashik","pune"],
    "marks":[91,92,93,94]
}
df=pd.DataFrame(data)
print(df)
encoded_df=pd.get_dummies(df,columns=["student","grades","course","city","marks"],dtype=int)
print(encoded_df)
```

```python
# ordinal Encoding 
from sklearn.preprocessing import OrdinalEncoder
size = [["l"],["M"],["s"],["x"],["xl"],["xxl"]]
encoder =  OrdinalEncoder(categories = [["l","M","s","x","xl","xxl"]])
encoded_size = encoder.fit_transform(size)
print(encoded_size)
```
